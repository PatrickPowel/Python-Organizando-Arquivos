```python
# LIMPANDO OS ARQUIVOS REMOVENDO EXTENSOES ['.jpg', '.jpeg', '.pdf', '.png', '.svg' '.docx', '.pdf', 'html'] 
# E IMPRIMINDO NA TELA QUAL ARQUVIO ESTÁ SENDO EXCLUIDO DENTRO DA PASTAS

import os

def excluir_arquivos_com_extensoes(diretorio, extensoes):
    for pasta_raiz, pastas, arquivos in os.walk(diretorio):
        for arquivo in arquivos:
            try:
                if any(arquivo.lower().endswith(ext) for ext in extensoes):
                    caminho_arquivo = os.path.join(pasta_raiz, arquivo)
                    os.remove(caminho_arquivo)
                    print(f'Arquivo excluído: {caminho_arquivo}')
            except PermissionError as e:
                print(f'Erro ao excluir o arquivo {arquivo}: {e}')
                # Adicionar condição para excluir a pasta se houver erro de permissão
                try:
                    os.rmdir(pasta_raiz)
                    print(f'Pasta excluída (se vazia): {pasta_raiz}')
                except OSError as e:
                    print(f'Erro ao excluir a pasta {pasta_raiz}: {e}')
            except Exception as e:
                print(f'Erro desconhecido ao processar o arquivo {arquivo}: {e}')

        # Verificar se a pasta está vazia e excluí-la se estiver
        if not os.listdir(pasta_raiz):
            try:
                os.rmdir(pasta_raiz)
                print(f'Pasta excluída (vazia): {pasta_raiz}')
            except OSError as e:
                print(f'Erro ao excluir a pasta vazia {pasta_raiz}: {e}')

# Substitua 'C:\Users\positivo\Desktop\arquivosSonia' pelo caminho do diretório que você deseja percorrer
# Substitua ['.jpg', '.jpeg', '.pdf', '.png'] pela lista de extensões que você deseja excluir
excluir_arquivos_com_extensoes(r"C:\Users\positivo\Desktop\arquivosSonia\matrizes2", ['.jpg', '.jpeg', '.pdf', '.png', '.svg', '.docx', '.pdf', '.rtf','.txt' 'html'])
```

```python
# EXTRAINDO PASTAS ZIPADAS, WINRAR E .LHA QUALQUER PASTAS QUE ESTEJA ZIPADA 
# O CODIGO EXTRAI NO MESMO DIRETÓRIO E DEPOIS EXCLUI A PASTA ZIP

import os
import zipfile
import rarfile
import lhafile

def extrair_arquivo(caminho_item):
    extensao = os.path.splitext(caminho_item)[1].lower()

    if extensao == '.zip':
        with zipfile.ZipFile(caminho_item, 'r') as arquivo_zip:
            arquivo_zip.extractall(os.path.dirname(caminho_item))
    elif extensao == '.rar':
        try:
            with rarfile.RarFile(caminho_item, 'r') as arquivo_rar:
                arquivo_rar.extractall(os.path.dirname(caminho_item))
        except rarfile.BadRarFile as e:
            print(f"Erro ao extrair {caminho_item}: {e}")
            return False  # Indica que ocorreu um erro
    elif extensao == '.lha':
        try:
            with lhafile.Lhafile(caminho_item) as arquivo_lha:
                arquivo_lha.extractall(os.path.dirname(caminho_item))
        except lhafile.BadLhafile as e:
            print(f"Erro ao extrair {caminho_item}: {e}")
            return False  # Indica que ocorreu um erro
    else:
        print(f"Arquivo {caminho_item} não é um arquivo ZIP, RAR ou LHA válido.")

    return True  # Indica que a extração foi bem-sucedida

def extrair_pasta(pasta):
    try:
        for item in os.listdir(pasta):
            caminho_item = os.path.join(pasta, item)

            try:
                if os.path.isfile(caminho_item) and (item.endswith('.zip') or item.endswith('.rar') or item.endswith('.lha')):
                    # Extrair o conteúdo do arquivo
                    sucesso = extrair_arquivo(caminho_item)

                    # Excluir o arquivo comprimido apenas se a extração for bem-sucedida
                    if sucesso:
                        os.remove(caminho_item)
                        print(f"Arquivo extraído e excluído com sucesso: {caminho_item}")
                elif os.path.isdir(caminho_item):
                    # Chamar a função recursivamente para subpastas
                    extrair_pasta(caminho_item)

                    # Verificar e excluir pastas vazias
                    if not os.listdir(caminho_item):
                        os.rmdir(caminho_item)
                        print(f"Pasta vazia excluída: {caminho_item}")
            except Exception as e:
                print(f"Erro ao processar {caminho_item}: {e}")
    except Exception as e:
        print(f"Erro ao processar pasta {pasta}: {e}")
        # Se houver erro ao processar a pasta, excluímos a pasta apenas se estiver vazia
        if not os.listdir(pasta):
            os.rmdir(pasta)
            print(f"Pasta com erro excluída: {pasta}")

if __name__ == "__main__":
    pasta_principal = r"C:\Users\positivo\Desktop\zipadas"
    print(f"Processando pasta principal: {pasta_principal}")
    extrair_pasta(pasta_principal)
    print("Processo concluído.")
```

```python
# Traz os arquivos para uma unica pasta
import os
import shutil

def agrupar_arquivos(caminho_origem, caminho_destino):
    # Cria a pasta de destino se não existir
    if not os.path.exists(caminho_destino):
        os.makedirs(caminho_destino)

    # Dicionário para rastrear nomes de arquivos e quantidades
    nomes_arquivos = {}

    # Percorre todas as pastas e subpastas
    for pasta_atual, _, arquivos in os.walk(caminho_origem):
        for arquivo in arquivos:
            # Cria o caminho completo do arquivo atual
            caminho_arquivo_atual = os.path.join(pasta_atual, arquivo)

            # Verifica se o nome do arquivo já existe no dicionário
            if arquivo in nomes_arquivos:
                # Adiciona um número ao final do nome do arquivo
                nome_base, extensao = os.path.splitext(arquivo)
                contador = nomes_arquivos[arquivo]
                novo_nome = f"{nome_base}_{contador}{extensao}"
                nomes_arquivos[arquivo] += 1
            else:
                # Mantém o nome original se não houver duplicatas
                novo_nome = arquivo
                nomes_arquivos[arquivo] = 1

            # Move o arquivo para a pasta de destino
            caminho_destino_arquivo = os.path.join(caminho_destino, novo_nome)

            try:
                shutil.move(caminho_arquivo_atual, caminho_destino_arquivo)
            except PermissionError as e:
                print(f"Erro ao mover '{caminho_arquivo_atual}': {e}")
                continue

    # Exclui pastas vazias
    for pasta_atual, subpastas, _ in os.walk(caminho_origem, topdown=False):
        for subpasta in subpastas:
            caminho_pasta_atual = os.path.join(pasta_atual, subpasta)
            if not os.listdir(caminho_pasta_atual):
                print(f'Excluindo pasta vazia: {caminho_pasta_atual}')
                try:
                    os.rmdir(caminho_pasta_atual)
                except PermissionError as e:
                    print(f"Erro ao excluir pasta vazia '{caminho_pasta_atual}': {e}")

if __name__ == "__main__":
    caminho_origem = r"C:\Users\positivo\Desktop\arquivosSonia\matrizes2"
    caminho_destino = r"C:\Users\positivo\Desktop\arquivosSonia\matrizes2"

    agrupar_arquivos(caminho_origem, caminho_destino)
    print('Operação concluída: Arquivos movidos para "arquivos-limpos"')
```

```python
# Extraindo arquivos de pastas e subpastas para qualquer tipo de pasta que seja zipada

import os
import shutil
import zipfile
import rarfile

def mover_arquivos_para_destino(origem, destino):
    # Percorre recursivamente os diretórios e subdiretórios
    for pasta_atual, _, arquivos in os.walk(origem):
        for arquivo in arquivos:
            caminho_origem = os.path.join(pasta_atual, arquivo)
            caminho_destino = os.path.join(destino, arquivo)

            # Verifica se o arquivo é um arquivo zip ou rar válido antes de mover
            if (arquivo.lower().endswith('.zip') or arquivo.lower().endswith('.rar')) and (zipfile.is_zipfile(caminho_origem) or rarfile.is_rarfile(caminho_origem)):
                # Move o arquivo para o destino
                shutil.move(caminho_origem, caminho_destino)

def excluir_pastas_vazias(diretorio):
    # Percorre os diretórios de trás para frente e exclui pastas vazias
    for pasta_atual, _, _ in os.walk(diretorio, topdown=False):
        try:
            os.rmdir(pasta_atual)
        except OSError:
            # Ignora se a pasta não estiver vazia
            pass

if __name__ == "__main__":
    # Defina o caminho da pasta de origem e de destino
    pasta_origem = os.getcwd()  # Pasta atual
    pasta_destino = r'C:\Users\positivo\Desktop\arquivosSonia\arquivos-limpos'

    # Move os arquivos para a pasta de destino
    mover_arquivos_para_destino(pasta_origem, pasta_destino)

    # Exclui pastas vazias no diretório de origem
    excluir_pastas_vazias(pasta_origem)

    print("Operação concluída.")
```

```python
import os
import shutil

def mover_arquivos_para_destino(origem, destino):
    try:
        # Cria a pasta de destino, se não existir
        os.makedirs(destino, exist_ok=True)

        # Percorre recursivamente os diretórios e subdiretórios
        for pasta_atual, _, arquivos in os.walk(origem):
            for arquivo in arquivos:
                caminho_origem = os.path.join(pasta_atual, arquivo)
                caminho_destino = os.path.join(destino, arquivo)

                # Verifica se é um arquivo antes de tentar mover
                if os.path.isfile(caminho_origem):
                    try:
                        # Move o arquivo para o destino
                        shutil.move(caminho_origem, caminho_destino)
                        print(f"Arquivo movido: {caminho_destino}")
                    except Exception as e:
                        print(f"Erro ao mover o arquivo {caminho_origem}: {e}")

    except Exception as e:
        print(f"Erro ao processar os arquivos: {e}")

if __name__ == "__main__":
    # Defina o caminho da pasta de origem e de destino
    pasta_origem = os.getcwd()  # Pasta atual
    pasta_destino = r'C:\Users\positivo\Desktop\arquivosSonia\arquivos-limpos'

    # Move os arquivos para a pasta de destino
    mover_arquivos_para_destino(pasta_origem, pasta_destino)

    print("Operação concluída.")

```

```python
import os
import shutil

caminho_base = r"C:\Users\positivo\Desktop\organizado"

# Lista para armazenar os caminhos dos arquivos
caminhos_arquivos = []

# Percorre todas as subpastas e coleta os caminhos dos arquivos
for pasta_raiz, _, arquivos in os.walk(caminho_base):
    caminhos_arquivos.extend([os.path.join(pasta_raiz, arquivo) for arquivo in arquivos])

# Organiza os caminhos dos arquivos por ordem alfabética
caminhos_arquivos.sort(key=lambda x: (not x[0].isdigit(), x))

# Cria um diretório de destino para os arquivos organizados
destino_organizado = os.path.join(caminho_base, "organizado")
os.makedirs(destino_organizado, exist_ok=True)

# Move os arquivos organizados para o novo diretório
for caminho_arquivo in caminhos_arquivos:
    nome_arquivo = os.path.basename(caminho_arquivo)
    novo_caminho = os.path.join(destino_organizado, nome_arquivo)
    
    try:
        shutil.move(caminho_arquivo, novo_caminho)
        print(f"Arquivo '{nome_arquivo}' movido para '{novo_caminho}'.")
    except FileNotFoundError as e:
        # Ignora erros relacionados à movimentação de arquivos/diretórios
        print(f"Ignorando erro: {e}")

print("Concluído!")

```
